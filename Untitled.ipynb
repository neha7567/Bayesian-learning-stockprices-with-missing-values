# Multiple Imputation of Missing Values in Stock Prices Data

## 1. Motivation

Financial stock market data, for various reasons, frequently contain missing
values. One reason for this is that, a value for stock is observed only when a trade order is executed from the limit order books and stock value for each minute is not always observed.
Investment banks are interested in this minute by minute stock price information as they use these prices to price other derivatives or hedge their risks from other riskier trades. Typically, one can use the last seen values of these stock prices or one could use other methods of imputation used in literature. One drawback of using the *last value carried forward* method is it assumes that prices remain unchanged when there is no data. The drawback of using simulations based imputation based methods is twofold - firstly, the time it takes to run multiple simulations can be more than the frequency of the data; secondly, more than 80 percent of the data has missing values. 
Further, it would be naive to assume that the missing values are *MCAR(missing completely at random)*. Under these circumstances, we assume that the missing data depends on the observed entries and we can use information of other similar stocks to form better estimates for the missing values.  
For instance, it can be seen from data that the stock prices for the five biggest tech firms, namely Apple, Facebook, Microsoft, Google and Amazon would move together in the same direction. 
The main goal of this study is to combine stock price data of these related stocks and develop an
algorithm to impute the missing values in individual stock prices using “information-sharing”
between different time series. To develop this imputation algorithm for the stock price time series data, we model the stock prices as a dynamic linear model. We compute the posterior distribution of the parameters of the model given the past data and simulate them using *Gibbs Sampler* to find the best estimate of the missing values.

## 3. Model  
We will now setup the linear dynmic model for the data. We assume that the stock price data $\mathbf{Y}_t$ can be given as :

\begin{eqnarray}
\mathbf{Y_t} = \theta_t + V_t
\end{eqnarray}

where, $\theta_t$ is the latent state variable. We assume that $V_t$ is a normally distributed with parameters $N(0, \Sigma_t)$. This implies that the observed $\mathbf{Y}_t \sim N(\theta_t, \Sigma_t)$. 
Since we donot observe the latent variable $\theta_t$, we have a prior that $\theta_t|_{\mathbf{y_1,\cdots \mathbf{y}_n}, \Sigma_t} \sim N(\mu_n,\Lambda_n)$

Further, we assume that $\Sigma_t \sim Inverse Wishart (\nu, \psi)$

Notice that $\mathbf{Y}_t$ is not fully observed, i.e at any given time time $t$, $\mathbf{Y}_t$ will have a a few missing stock prices $\mathbf{Y}_{mis, t}$ and few observed prices $ \mathbf{Y}_{obs, t}$. Our goal is to jointly estimate $\Sigma_t, \theta_t, \mathbf{Y}_{mis,t}$.

Let the prior distribution of the variables be given as 

\begin{eqnarray}
&\theta_t \sim N(\mu_0, \Lambda_0)\\
&\Sigma_t \sim InvWishart(\mu_0, \psi_0)
\end{eqnarray}

Then, we can update these distributions using data and from Hoff 2007, we have that the posterior distribution can given as :

\begin{eqnarray}
&\theta_n|_{\mathbf{y_1,\cdots \mathbf{y}_n}, \Sigma_t} \sim N(\mu_n, \Lambda_n)\\
&\Sigma_n|_{\mathbf{y_1,\cdots \mathbf{y}_n}, \theta_t} \sim InvWishart(\nu_0 + n, \psi_n)\\
&\mathbf{Y}_{mis, n}|_{\theta_t, Sigma_t,\mathbf{y_1,\cdots \mathbf{y}_{obs,n}} } \sim N(\theta_p, \Sigma_p)
\end{eqnarray}


where, 
\begin{eqnarray}
& \mu_n = (\Lambda_o{}^{-1} + n \sigma_n{}^{-1})^{-1} (\Lambda_o{}^{-1} \mu_0 + n \sigma_n{}^{-1}\bar{\mathbf{y}} )   & \Lambda_n = (\Lambda_0{}^{-1} + n \sigma_n{}^{-1})^{-1}\\
& \psi_n = (\psi_0 + \sum_{i=1}^{n}(\mathbf{y}_i - \theta_n)(\mathbf{y}_i - \theta_n)^{T})^{-1} & \\
& \theta_p = \theta_{mis,n} + \Sigma_{mis,obs,n} (\Sigma_{obs,obs,n})^{-1}(\mathbf{y}_{obs,n} - \theta_{obs,n})& \Sigma_p = Sigma_{mis,mis,n} - \Sigma_{mis,obs,n} (\Sigma_{obs,obs,n})^{-1} \Sigma_{obs,mis,n}
\end{eqnarray}   

## 4 Methodology and Model Validation

While $\Sigma, \theta$ are unknown and need to be updated using $\mathbf{Y}$, notice that $\mathbf{Y}$ itself is not completely observed. While we know the *full conditional distribution* of each of these quantities, we need Gibbs sampler to have an estimate of the joint probability. 

Furthermore, since the true values are never observed, we test our algorithm on dataset we generate. More specifically we generate a stock price data set using Geometric brownian motion, we then randomly delete some of the data entries of the data. Finally we use our bayesian update with Gibbs sampling to estimate the missing values. Finally we will compare the **Mean Squared Error** of our algorithm with the *last value carried forward*

We will now describe our algorithm, then generate the dataset and find the Mean squared error from our algorithm. 

### 4.1 Algorithm 

1. At each given time $t$, impute some value for the $\mathbf{Y}_{mis,t}$. In our algorithm, we set $\mathbf{Y}_{mis,t}$ to the mean value from last $T$ values.
2. Choose an initial value of $\Sigma_t$. We choose this by looking back $T$ time periods of imputed dataset and finding the covariance matrix. 
3. Setup the Gibbs sampler with 1000 iterations. 

a. Sample $\theta$ from $N(\mu_n, \Lambda_n)$ where $\mu_n, \Lambda_n $ are computed using the initial values of $\sigma$ and $\mathbf{Y}$.

b. Using this sample $\theta$, compute $\psi_n$ and sample $\Sigma$ from $InvWishart(\nu_0 + n, \psi_n)$.

c. Finally, sample $\mathbf{Y}_{mis, n}$ from $N(\theta_p, \Sigma_p)$

### 4.2 Genearting Mock dataset 

```python
import random 

def gen_paths(S0, r, sigma, T, M, I): # function that generates a geometric brownian motion 
    dt = float(T) / M
    paths = np.zeros((M + 1, I), np.float64)
    paths[0] = S0
    for t in range(1, M + 1):
        rand = np.random.standard_normal(I)
        paths[t] = paths[t - 1] * np.exp((r - 0.5 * sigma ** 2) * dt +
                                         sigma * np.sqrt(dt) * rand)

    return paths

def generate_incomplete_data(T):
    data = {'apple': gen_paths(50, .50, .15, 1, T, 100)[:, -1].tolist(),
            'facebook': gen_paths(100, .65, .25, 1, T, 100)[:, -1].tolist(),
            'google': gen_paths(75, .25, .34, 1, T, 100)[:, -1].tolist(),
            'amazon': gen_paths(30, .3, .24, 1, T, 100)[:, -1].tolist(),
            'microsoft': gen_paths(45, .35, .5, 1, T, 100)[:, -1].tolist()}
    path = pd.DataFrame(data, columns=['apple', 'facebook', 'google', 'amazon', 'microsoft'])
    pickle.dump(path, open('data_mock_complete.p','wb'))

    for rows in range(len(path)):
        num_missing = random.randrange(0, 4)
        missing_index = random.sample(list(range(0, 5)), num_missing)  # return [] when num_missing is 0
        cols = ['apple', 'facebook', 'google', 'amazon', 'microsoft']
        if missing_index is not None:
            for i in missing_index:
                path[cols[i]].iloc[rows] = 0
    pickle.dump(path, open('data_mock_incomplete.p','wb'))
    return path

generate_incomplete_data(T= 300)
```

```python
import scipy.stats as stats

data_mock = pickle.load(open('data_mock_incomplete.p', 'rb'))
data_mock_na = data_mock.replace(0, np.nan)
data_mock_ff = data_mock_na.fillna(method='ffill')
data_new = pd.DataFrame(columns=['apple', 'facebook', 'google', 'amazon', 'microsoft', 'theta', 'sigma'])
#data_new[['apple', 'facebook', 'google', 'amazon', 'microsoft']] = \
    #data_ff[['apple', 'facebook', 'google', 'amazon', 'microsoft']].copy(deep=True)

total_index_arr = np.array([0, 1, 2, 3, 4])
look_back = 10

initial_mean = data_mock_ff.mean(axis=0).to_numpy()
initial_lambda_inv = 20 * np.diag(np.ones(5))

data_mock_ff = data_mock_ff.fillna(value= {'apple':initial_mean[0], 'facebook':initial_mean[1],
                                      'google':initial_mean[2], 'amazon': initial_mean[3], 'microsoft':initial_mean[4]})

deg_free = 10
iter_num = 1000

# for each t find the best estimate
for index, row in data_mock.iterrows():
    look_back_list = [max(index - i, 0) for i in range(1, look_back + 1)]
    arr = np.array(look_back_list)
    look_back_list = np.unique(arr).tolist()
    look_back_period = len(np.unique(arr))

    y = row.to_numpy()

    non_zero_index = np.where(y)[0]
    zero_index = np.setdiff1d(total_index_arr, non_zero_index)

    # initial y estimate estimate

    y = data_mock_ff.iloc[index].to_numpy()

    relevant_df = pd.DataFrame(columns=['apple', 'facebook', 'google', 'amazon', 'microsoft'])
    relevant_df = relevant_df.append(data_mock_ff.iloc[look_back_list])

    # initial sigma  estimate estimate

    if index < 2:
        sigma = 20 * np.diag(np.ones(5))
    else:
        sigma = relevant_df.cov().to_numpy()

    theta = np.zeros(len(data_mock_ff.iloc[1]))

    theta_sim = 0
    sigma_sim = 0
    y_sim = 0

    for i in range(iter_num):
        # starting gibbs sampler
        y_sim += y
        new_row = {'apple': y[0], 'facebook': y[1], 'google': y[2], 'amazon': y[3], 'microsoft': y[4]}

        relevant_df = relevant_df.append(new_row, ignore_index=True)

        mean_y_look_back = relevant_df.mean().to_numpy()

        relevant_df = relevant_df[:-1]

        # sample theta
        a = np.linalg.pinv(initial_lambda_inv + look_back_period * np.linalg.pinv(sigma))
        b = np.matmul(initial_lambda_inv, initial_mean) + look_back_period * np.matmul(np.linalg.pinv(sigma),
                                                                                       mean_y_look_back)
        mean_theta_look_back = np.matmul(a, b)
        var_theta_look_back = np.linalg.pinv(initial_lambda_inv + look_back_period * np.linalg.pinv(sigma))
        theta = stats.multivariate_normal.rvs(mean_theta_look_back, var_theta_look_back,size=100)
        theta = theta.mean(axis=0)

        theta_sim += theta

        # using theta, sample sigma
        s_theta = np.matmul(np.transpose(y - theta), (y - theta))
        sigma = stats.invwishart.rvs(deg_free + look_back_period,
                                     np.linalg.pinv(s_theta + np.linalg.pinv(initial_lambda_inv)),size=100)
        sigma = sigma.mean(axis=0)

        sigma_sim += sigma

        if zero_index.size:
            # sample y_missing using theta, sigma
            sigma_b = sigma[zero_index, :]
            sigma_a = sigma[non_zero_index, :]
            sigma_b_a = sigma_b[:, non_zero_index]
            sigma_a_b = sigma_a[:, zero_index]
            sigma_a_a = sigma_a[:, non_zero_index]
            sigma_b_b = sigma_b[:, zero_index]
            sigma_mul = np.matmul(sigma_b_a, np.linalg.pinv(sigma_a_a))
            theta_p = theta[zero_index] + np.matmul(sigma_mul, np.subtract(y[non_zero_index], theta[non_zero_index]))
            sigma_p = sigma_b_b - np.matmul(sigma_mul, sigma_a_b)
            y_mis = stats.multivariate_normal.rvs(theta_p, sigma_p,size=100)
            y_mis = y_mis.mean(axis=0)
            y[zero_index] = y_mis

    theta = theta_sim/iter_num
    sigma = sigma_sim/ iter_num
    y = y_sim/iter_num

    row = {'apple': y[0], 'facebook': y[1], 'google': y[2], 'amazon': y[3], 'microsoft': y[4], 'theta': theta,
           'sigma': sigma}
    data_new = data_new.append(row, ignore_index=True)

pickle.dump(data_new, open('data_mock_imputed.p', 'wb'))
```

## 4. Results 

Now, that we have both the imputed values and last value carried forward, we will compare the mean squared error of both values. 

```python
imputed_data_mock = pickle.load(open('data_mock_imputed.p', 'rb'))
imputed_data_mock = imputed_data_mock.drop(['theta', 'sigma'], axis=1)
complete_data_mock = pickle.load(open('data_mock_complete.p', 'rb'))
for col in file_list:
    plt.plot(complete_data_mock[col], label='actual')
    plt.plot(data_mock_ff[col], label = 'ff')
    plt.plot(imputed_data_mock[col],label = 'imputed')
    plt.legend()
    plt.show()

error_lcf = complete_data_mock.subtract(data_mock_ff)
error_imputed = complete_data_mock.subtract(imputed_data_mock)

plt.plot(error_lcf, label='lcf')
plt.plot(error_imputed, label='imputed')
plt.legend()
plt.show()

error_lcf = error_lcf**2
error_lcf = (error_lcf.values.sum())**.5
error_imputed = error_imputed**2
error_imputed = (error_imputed.values.sum())**.5
print('error lcf='+str(error_lcf)+'and error imputed'+ str(error_imputed))
```

## 5. Implementation on Real data 

```python
data = pickle.load(open('real_data_original.p', 'rb'))
data = data.reset_index()
data = data.drop(['Time'],axis=1)
# remove the initial data where everything is nan
start_id = 0
for i, r in data.iterrows():
    if not(r.isna().values.all(axis=0)):
        start_id = i
        break

data = data.iloc[start_id:]
data = data.reset_index()
data = data.drop(['index'],axis=1)

time_series_ff = time_series_ff.drop(['level_0','index'],axis=1)
```

```python

data_new = pd.DataFrame(columns=['apple', 'google', 'microsoft','facebook', 'amazon', 'theta', 'sigma'])


total_index_arr = np.array([0, 1, 2, 3, 4])
look_back = 10

initial_mean = time_series_ff.mean(axis=0).to_numpy()
initial_lambda_inv = 20 * np.diag(np.ones(5))

time_series_ff = time_series_ff.fillna(value= {'apple':initial_mean[0], 'google':initial_mean[1],
                                      'microsoft':initial_mean[2], 'facebook': initial_mean[3],
                                               'amazon':initial_mean[4]})

deg_free = 10
iter_num = 100

# for each t find the best estimate
for index, row in data.iterrows():
    look_back_list = [max(index - i, 0) for i in range(1, look_back + 1)]
    arr = np.array(look_back_list)
    look_back_list = np.unique(arr).tolist()
    look_back_period = len(np.unique(arr))

    y = row.to_numpy()

    non_zero_index = np.where(y)[0]
    zero_index = np.setdiff1d(total_index_arr, non_zero_index)

    # initial y estimate estimate

    y = time_series_ff.iloc[index].to_numpy()

    relevant_df = pd.DataFrame(columns=['apple', 'google', 'microsoft','facebook', 'amazon'])
    relevant_df = relevant_df.append(time_series_ff.iloc[look_back_list])

    # initial sigma  estimate estimate

    if index < 2:
        sigma = 20 * np.diag(np.ones(5))
    else:
        sigma = relevant_df.cov().to_numpy()

    theta = np.zeros(len(data_mock_ff.iloc[1]))

    theta_sim = 0
    sigma_sim = 0
    y_sim = 0

    for i in range(iter_num):
        # starting gibbs sampler
        y_sim += y
        new_row = {'apple': y[0], 'google': y[1], 'microsoft': y[2], 'facebook': y[3], 'amazon': y[4]}

        relevant_df = relevant_df.append(new_row, ignore_index=True)

        mean_y_look_back = relevant_df.mean().to_numpy()

        relevant_df = relevant_df[:-1]

        # sample theta
        a = np.linalg.pinv(initial_lambda_inv + look_back_period * np.linalg.pinv(sigma))
        b = np.matmul(initial_lambda_inv, initial_mean) + look_back_period * np.matmul(np.linalg.pinv(sigma),
                                                                                       mean_y_look_back)
        mean_theta_look_back = np.matmul(a, b)
        var_theta_look_back = np.linalg.pinv(initial_lambda_inv + look_back_period * np.linalg.pinv(sigma))
        theta = stats.multivariate_normal.rvs(mean_theta_look_back, var_theta_look_back,size=100)
        theta = theta.mean(axis=0)

        theta_sim += theta

        # using theta, sample sigma
        s_theta = np.matmul(np.transpose(y - theta), (y - theta))
        x = np.linalg.pinv(s_theta + np.linalg.pinv(initial_lambda_inv))
        try:
            np.linalg.cholesky(x)
        except np.linalg.LinAlgError:
            eigen_val, eigen_vector = np.linalg.eig(x)
            eigen_val[eigen_val <= 0] = 1
            x = np.diag(eigen_val)
        
        
        sigma = stats.invwishart.rvs(deg_free + look_back_period, x, size=100)
        sigma = sigma.mean(axis=0)

        sigma_sim += sigma

        if zero_index.size:
            # sample y_missing using theta, sigma
            sigma_b = sigma[zero_index, :]
            sigma_a = sigma[non_zero_index, :]
            sigma_b_a = sigma_b[:, non_zero_index]
            sigma_a_b = sigma_a[:, zero_index]
            sigma_a_a = sigma_a[:, non_zero_index]
            sigma_b_b = sigma_b[:, zero_index]
            sigma_mul = np.matmul(sigma_b_a, np.linalg.pinv(sigma_a_a))
            theta_p = theta[zero_index] + np.matmul(sigma_mul, np.subtract(y[non_zero_index], theta[non_zero_index]))
            sigma_p = sigma_b_b - np.matmul(sigma_mul, sigma_a_b)
            y_mis = stats.multivariate_normal.rvs(theta_p, sigma_p,size=100)
            y_mis = y_mis.mean(axis=0)
            y[zero_index] = y_mis

    theta = theta_sim/iter_num
    sigma = sigma_sim/ iter_num
    y = y_sim/iter_num

    row = {'apple': y[0], 'google': y[1], 'microsoft': y[2], 'facebook': y[3], 'amazon': y[4], 
           'theta': theta, 'sigma': sigma}
    data_new = data_new.append(row, ignore_index=True)

pickle.dump(data_new, open('data_imputed.p', 'wb'))
```
